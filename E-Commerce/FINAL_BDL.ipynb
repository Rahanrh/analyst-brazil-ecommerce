Melakukan Visualisasi Data setelah melakukan Clean Data

```python
import matplotlib.pyplot as plt
import seaborn as sns
total = len(dataSET_clean)
ax1 = plt.figure(figsize=(23,7))

g = sns.countplot(x='customer_state', data=dataSET_clean)
g.set_title("Konsumen berdasar State", fontsize=20)
g.set_xlabel("State", fontsize=17)
g.set_ylabel("Jumlah", fontsize=17)
sizes = []
for p in g.patches:
    height = p.get_height()
    sizes.append(height)
    g.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:1.2f}%'.format(height/total*100),
            ha="center", fontsize=15) 
g.set_ylim(0, max(sizes) * 1.1)
```
